```python
"""
[메인 퀘스트 1번: 은행계좌 만들기 - 3점]
Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후
생성자(hint: 매직매서드..!!)를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는
3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)
Q2. 클래스 변수: 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장하세요.
Q3. 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요.
Q4. 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다.
Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.
Q6. 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를
출력하세요.
(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
Q7. 이자 지급하기 : 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.
Q8. 여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.
Q9. 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.
Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
(입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)
"""
import random

class Account:

    acc_cnt = 0   #계좌 객체의 개수(클래스 변수)

    #Account 클래스를 생성한 후
    #은행이름, 예금주, 계좌번호, 잔액이 설정
    def __init__(self, user_name, balance):

        self.deposit_cnt = 0 #입금횟수 초기화
        self.deposit_hist = []  #추후 입금내역이 기록정보를 출력할때 이용
        self.withdraw_hist = [] #추후 출금내역이 기록정보를 출력할때 이용

        # 입력파라메터: 예금주,초기잔액(balance)
        self.user_name = user_name
        self.balance =  balance

        # 정보세팅: 은행이름: SC은행, 계좌번호: 111-11-111111)로 랜덤하게 생성
        self.bank = 'SC은행'  #은행이름

        #계좌번호: 111-11-111111 로 랜덤하게 생성 (생성값은 int)
        acc_num1 = random.randint(0, 999) #111-
        acc_num2 = random.randint(0, 99)  #-11-
        acc_num3 = random.randint(0, 999999)  #-111111

        #계좌번호 -> 생성된값은 int -> str로 변경하고 중간에 '-'값을 위치에 맞게 수정후 계좌설정하기
        self.account_num = str(acc_num1).zfill(3) + '-' + str(acc_num2).zfill(2)  + '-' + str(acc_num3).zfill(6)

        Account.acc_cnt += 1  #생성자가 호출된 횟수 count


    #생성된 계좌의 개수를 출력
    def get_account_num():
        print("생성된 계좌의 총 개수:"+ str(Account.acc_cnt))

    #정보 출력 메서드-저장된 정보를 출력, 잔고는 세자리마다 쉼표 출력
    #(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
    def display_info(self):
        print("은행이름:", self.bank, ", 예금주:", self.user_name, ", 계좌번호:", self.account_num, ", 잔고:", format(self.balance, ','), "원")


    #입금메서드, 입금은 최소 1원 이상만 가능
    def deposit(self, amount):  #입금액

        if amount >= 1: #입금액이 1원이상인 경우에만 처리
            self.balance += amount  #잔고에 입력한 입금액을 더한다.
            #print("잔고:"+str(self.balance),"입금액:", str(amount))

            self.deposit_hist.append(amount) #이력list에 입금액 정보를 추가

            self.deposit_cnt += 1 #정상적으로 입금되면 입금횟수 1증가
            self.deposit_hist.append(self.deposit_cnt) #횟수

            self.bal_amount_nm = ''
            #이자지급하기 (입금 횟수가 5회가 될 때, 잔고를 기준으로 1%의 이자가 잔고에 추가)
            if self.deposit_cnt % 5 == 0: #5의 배수이면(5,10,15,20...)
                self.balance =  (self.balance * 1.01) #1%의 증가한 이자값
                self.bal_amount_nm = "이자지급"
                print(str(self.balance)+"원의 이자가 지급되었습니다.")
            else:
                self.bal_amount_nm = "입금"

            self.deposit_hist.append(self.bal_amount_nm) #입급/이자지급 여부
            self.deposit_hist.append(self.balance) #잔액

        else:
            print("입금은 최소 1원 이상만 가능합니다.")


    #입금내역이 기록
    def deposit_history(self):
        #for self.amount in self.deposit_hist:
            #print("입금액: " + format(self.amount, ','), "잔고: " + format(self.balance, ','))

        for i in range(len(self.deposit_hist)):
            print(self.deposit_hist[i])


    #출금 메서드, 출금은 계좌의 잔고 이상으로 출금불가
    def withdraw(self, amount): #출금액을 받아서  체크
        if amount <= self.balance:
            self.balance -= amount
            #print("잔고:"+str(self.balance),"출금액:",str(amount))

            self.withdraw_hist.append(amount) #이력list에 출금액 정보를 추가
        else:
            print("계좌 잔고 이상으로 출금할 수 없습니다.")

    #출금 내역을 출력
    def withdraw_history(self):
        for self.amount in self.withdraw_hist:
            print("출금액: " + format(self.amount, ','), "잔고: " + format(self.balance, ','))


#main 실행시 여러 객체 생성(최소3개이상), 생성된 인스턴스를 리스트에 저장
#객체 순회 반복문을 통해, 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력

data_list = []

a1 = Account("노은하", 1)
b1 = Account("김세로이", 100)
c1 = Account("하늬", 2)

data_list.append(a1)
data_list.append(b1)
data_list.append(c1)

#생성된 계좌의 개수를 출력
Account.get_account_num()

a1.display_info()
b1.display_info()
c1.display_info()

a1.deposit(300000)
a1.deposit(100000)
a1.deposit(20000)
a1.deposit(400000)
a1.deposit(500000)
a1.deposit_history()
a1.withdraw(23000)
a1.withdraw_history()


#생성된 계좌의 개수를 출력
b1.deposit(5000000)
b1.deposit(10000)
b1.deposit(20000)
b1.deposit_history()

b1.withdraw(1200000)
b1.withdraw_history()

#생성된 계좌의 개수를 출력
c1.deposit(120456)
c1.deposit(2000)
c1.deposit(30000)
c1.deposit_history()
c1.withdraw(500000)
c1.withdraw(1000)
c1.withdraw_history()


cnt1 = 0
for ix in data_list:
    if ix.balance >= 1000000: #잔고가 100만원 이상인 고객

        if cnt1 == 0:
            print("\n===== 잔고가 100만원 이상인 고객리스트 ======")

        ix.display_info()
        cnt1 += 1
```
